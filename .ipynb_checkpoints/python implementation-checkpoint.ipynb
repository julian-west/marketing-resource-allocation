```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
import re

import statsmodels.api as sm

from sklearn.model_selection import train_test_split

plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
```

# Load Data

```python
data = pd.read_excel("./data/beer_data_chicago_Dominicks.xlsx")
```

# EDA

```python
BRAND_COLORS = ["#00AFBB", "#E7B800", "#ff0000"]

def get_cols(regex_pat, columns=data.columns):
    return [col for col in columns if re.match(regex_pat,col)]

def plot_cols(ax, cols, title, ylabel, xlabel=None, data=data):
    data[cols].plot(color=BRAND_COLORS, ax=ax)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    
    ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    
    return ax
```

```python
brand_sales_cols = get_cols(r"^SALES")
brand_price_cols = get_cols(r"^PRICE")
brand_display = get_cols(r"display")
brand_feature = get_cols(r"FEATURE")


fig, axs = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

axs[0][0] = plot_cols(axs[0][0],brand_sales_cols, "Sales by brand", "Sales $")
axs[0][1] = plot_cols(axs[0][1],brand_price_cols, "Brand price trend", "Price $")
axs[1][0] = plot_cols(axs[1][0],brand_display, "Brand display", "% of SKUs")
axs[1][1] = plot_cols(axs[1][1],brand_feature, "Brand feature", "% of SKUs")


plt.subplots_adjust(hspace=0.3)
plt.show()
```

**Split into test and train**

```python
def create_modelling_df(data, brand):
    """Create dataframe with transformed variables for modelling

    Args:
        data (pd.DataFrame): raw data loaded from Excel
        brand (int or None): the brand for which we are predicting sales (e.g. 1,2 or 3). If None, 
            model will use total category sales

    Returns:
        modelling_data (pd.DataFrame): dataframe for modelling with target column (log brand sales) 
            and features
    """

    # add month variable
    data['date'] = pd.date_range(start='1989/09/01', periods=len(data), freq='W')
    data['month'] = data['date'].dt.month_name()

    # create seasonal dummies (drop one month)
    seasonal_dummies = pd.get_dummies(data['month'], drop_first=True)

    # columns to log transform (sales, display and price)
    log_cols = [col for col in data.columns if re.match(
        "CategorySales|SALES|display|^PRICE", col)]
    # columns representing brand 'Feature'
    feature_cols = [col for col in data.columns if 'FEATURE' in col]

    # create dataframe for modelling
    modelling_data = pd.DataFrame()

    # populate modelling dataframe
    for col in log_cols:
        # apply log transformation
        modelling_data[f"log_{col}"] = np.log(data[col]+1)

    for col in feature_cols:
        # make featured binary
        modelling_data[f"binary_{col}"] = data[col].apply(lambda x: 1 if x > 0.01 else 0)
    
    #add seaonsal column
    modelling_data[seasonal_dummies.columns] = seasonal_dummies

    if brand == None:
        target_col = 'log_CategorySales'
    else:
        target_col = f'log_SALESBRAND{brand}'

    final_cols = [
        col for col in modelling_data.columns if 'SALES' not in col and 'CategorySales' not in col]

    return modelling_data[final_cols + [target_col]]


def create_train_test_split(data, train_size=0.75, seed=42):
    """Create the training and test datasets

    Randomly shuffle datapoints

    Args:
        data (pd.DataFrame): modelling data for a particular brand. The target column should be
            the final column in the dataframe
    """
    return train_test_split(data.iloc[:, :-1], data.iloc[:, -1], shuffle=True,
                            train_size=train_size, random_state=seed)
```

**Create dataframes for each brand**

```python
brand1_df = create_modelling_df(data=data, brand=1)
brand2_df = create_modelling_df(data=data, brand=2)
brand3_df = create_modelling_df(data=data, brand=3)
brandAll_df = create_modelling_df(data=data, brand=None)
```

```python
def estimate_ols(X,y):
    """Estimate the OLS model
    
    If 'validate' is true, the function will return the OLS regression models when run on a training
    and validation set.
    
    Args:
        data (pd.DataFrame): output from `create_modelling_df` function
        
    Returns:
        results (fitted regression model)
    
    """
    X = sm.add_constant(X)
    model = sm.OLS(y,X)
    return model.fit()    
    
```

```python
from sklearn.metrics import mean_squared_error

X_train_b1, X_test_b1, y_train_b1, y_test_b1 = create_train_test_split(brand1_df)

b1_model_results = estimate_ols(X_train_b1,y_train_b1)

X = sm.add_constant(X_test_b1)
preds = b1_model_results.predict(X)
mean_squared_error(preds, y_test_b1)
```

```python
print(b1_model_results.summary())
```

```python
X_train_all, X_test_all, y_train_all, y_test_all = create_train_test_split(brandAll_df)

b1_model_results = estimate_ols(X_train_all,y_train_all)

X = sm.add_constant(X_test_all)
preds = b1_model_results.predict(X)
mean_squared_error(preds, y_test_all)
```
